```python
estudiantes["extra"] = estudiantes["absences"] * 100 / 93

estudiantes


```

```python

def approved(row):
    if row["extra"] < 80:
        return "Reprobado"
    elif row["extra"] >= 80 and row["G3"] < 10:
        return "Reprobado"
    elif row["extra"] >= 80 and row["G3"] >= 10 and row["G3"] <= 15:
        return "Aprobado"
    elif row["extra"] >= 80 and row["G3"] > 15:
        return "Aprobado"

estudiantes["approved"] = estudiantes.apply(approved, axis=1)
estudiantes

```
